```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

athletes_2020_df = pd.read_csv('./Resources/athletic_sales_2020.csv')

athletes_2021_df = pd.read_csv('./Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame

athletes_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
athletes_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.

athletes_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletes_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

combined_athletes_df = pd.concat([athletes_2020_df, athletes_2021_df], ignore_index=True)
```

```python
# Check if any values are null.

combined_athletes_df.isnull().sum()
```

```python
# Check the data type of each column

combined_athletes_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype

combined_athletes_df['invoice_date'] = pd.to_datetime(combined_athletes_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.

combined_athletes_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

combined_athletes_df.groupby(['region', 'state', 'city'])['product'].count().reset_index(name='Total_Products_Sold')


# Show the top 5 results.

combined_athletes_df.groupby(['region', 'state', 'city'])['product'].count().reset_index(name='Total_Products_Sold').head()
```

#### Using `pivot_table`

```python
# Create a pivot table to show the number of products sold for each region, state, and city
pivot_retailer_sales = combined_athletes_df.pivot_table(
    values='units_sold', 
    index=['region', 'state', 'city'], 
    aggfunc='sum'
).reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_retailer_sales.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results
pivot_retailer_sales.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

total_sales_df = combined_athletes_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index(name='Total_Sales')
# Show the top 5 results.

total_sales_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

pivot_total_sales_df = combined_athletes_df.pivot_table(
    values='total_sales', 
    index=['region', 'state', 'city'], 
    aggfunc='sum'
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"

pivot_total_sales_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.

pivot_total_sales_df.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"

retailer_sales_df = combined_athletes_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index(name='Total_Sales')

# Show the top 5 results.

retailer_sales_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.

pivot_retailer_sales_df = combined_athletes_df.pivot_table( values='total_sales', 
    index=['retailer','region', 'state', 'city'], 
    aggfunc='sum'
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"

pivot_retailer_sales_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the top 5 results.

pivot_retailer_sales_df.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the DataFrame for Women's Athletic Footwear
women_footwear_sales_df = combined_athletes_df[combined_athletes_df['product'] == "Women's Athletic Footwear"]

# Show the top 5 results.

women_footwear_sales_df.head()
```

#### Using `groupby`

```python
# using groupby to show the Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

total_women_footwear_sales_df = combined_athletes_df.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index(name='Womens_Footwear_Units_Sold')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

total_women_footwear_sales_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.

total_women_footwear_sales_df.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

pivot_women_footwear_sales_df = combined_athletes_df.pivot_table(
    values='units_sold', 
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum'
).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

pivot_women_footwear_sales_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.

pivot_women_footwear_sales_df.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pivot_date_sales_df = combined_athletes_df.pivot_table(index='invoice_date', values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"

pivot_date_sales_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)

# Show the table.

pivot_date_sales_df
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample to weekly bins for total sales
weekly_sales_df = pivot_date_sales_df.resample('W').sum()

# Sort the data in descending order
sorted_weekly_sales = weekly_sales_df.sort_values(by='Total_Sales', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(weekly_sales_df.index, weekly_sales_df['Total_Sales'], marker='o')
plt.title('Weekly Total Sales of Womens Athletic Footwear')
plt.xlabel('Week')
plt.ylabel('Total Sales')
plt.show()
```
